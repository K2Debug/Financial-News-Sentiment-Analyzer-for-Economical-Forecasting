# EF-02 Documentation: Pipeline Overview

**Project:** Tanzania Financial News Sentiment Analyser  
**Programme:** BSc Data Science and AI, Ardhi University  
**Author:** K2 | Group 8

---

## Research question

Does the tone of Tanzanian financial news headlines correlate with measurable macroeconomic outcomes — specifically TZS inflation and the USD/TZS exchange rate — at the monthly level (2022–2024)?

---

## Two layers in this project

| Layer | Location | Purpose |
|-------|----------|--------|
| **Production pipeline** | `../notebooks/01_`–`06_` | Execution-only notebooks that scrape, clean, classify, merge, and visualise data |
| **Documentation** | `EF-02/docs/` (this folder) | Markdown-heavy walkthroughs explaining *how* each feature works, with small demos |

**Start here:** [`00_setup.ipynb`](00_setup.ipynb) — install packages section by section (FinBERT / Transformers is one large optional section).

You can read the other docs in any order after setup. To reproduce the full dataset, run the production notebooks in numeric order.

**Tip:** All `demo_*.ipynb` notebooks are progressive tutorials — start at Part 1 / Section 1 and work through in order. The scraper doc ends with a CSS selector guide for building your own config.

---

## Pipeline flow

```
01_scraper  →  02_cleaning  →  03_benchmarking  →  04_classifier
                                                      ↓
06_visualisation  ←  05_consolidation  ←  labelled headlines + CPI + FX
```

**Note:** Benchmarking (`03_`) was run *before* scraping in the real project — it proved LLMs could classify Tanzanian financial headlines on a labelled sample before investing time in a full scrape.
