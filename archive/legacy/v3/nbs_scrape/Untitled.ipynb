```python
import pandas as pd

path = "en-1736337764-CPI Summary_122024.xls"
sheets = {
    '2022': '2022_REBASED SERIES',
    '2023': '2023_REBASED SERIES ',
    '2024': '2024_REBASED SERIES  ',
}

records = []

```

```python
for year, sheet_name in sheets.items():
    df = pd.read_excel(path, sheet_name=sheet_name, engine='xlrd', header=None)
    
    # Row 1 has the month columns (as datetime), cols 3 onward
    header_row = df.iloc[1]
    months = header_row[4:16].tolist()  # 12 months + skip annual avg
    
    # Row 2 = Inflation Rate, Row 3 = All Items Index
    inflation_row = df.iloc[2, 3:15].tolist()
    index_row = df.iloc[3, 3:15].tolist()
    
    for i, month in enumerate(months):
        records.append({
            'date': pd.to_datetime(month),
            'inflation_rate_pct': inflation_row[i],
            'all_items_index': index_row[i],
        })

cpi_df = pd.DataFrame(records)
cpi_df = cpi_df[cpi_df['date'].dt.year.isin([2022, 2023, 2024])]
cpi_df = cpi_df.sort_values('date').reset_index(drop=True)

print(cpi_df)
```

```python
cpi_df.to_csv('tanzania_cpi_2022_2024.csv', index=False)
```
